```python
import pandas as pd
from datasets import DatasetDict, Dataset, load_from_disk, load_metric, load_dataset
from konlpy.tag import *
```

```python
mecab = Mecab()
```

```python
dataset = load_from_disk("/content/data/train_dataset")
test_dataset=load_from_disk("/content/data/test_dataset")

train_dataset = dataset["train"]
valid_dataset = dataset["validation"]
test_dataset=test_dataset['validation']

train_df=pd.DataFrame(train_dataset)
valid_df=pd.DataFrame(valid_dataset)
test_df=pd.DataFrame(test_dataset)
```

```python
train_df.head()
```

```python
mecab.tagset
```

```python
mecab.pos('지구에서 가장 큰 새는?')
```

```python
def noun_adj(txt):
  return ", ".join([word for word, tag in mecab.pos(txt) if tag in ['NNG', 'NNP', 'VA', 'VA+ETM']]) +" "+ txt # 일반 명사, 고유명사, 형용사, 관형사
```

```python
train_df['question'] = train_df['question'].apply(noun_adj)
```

```python
valid_df['question'] = valid_df['question'].apply(noun_adj)
```

```python
test_df['question'] = test_df['question'].apply(noun_adj)
```

```python
test_df
```

```python
train_dataset_aug = Dataset.from_pandas(train_df)
valid_dataset_aug = Dataset.from_pandas(valid_df)
datasets_aug = DatasetDict({'train': train_dataset_aug, 'validation': valid_dataset_aug})
test_dataset_aug = Dataset.from_pandas(test_df)
test_dataset_aug_dict = DatasetDict({'validation': test_dataset_aug})
```

```python
datasets_aug.save_to_disk("noun_adj_aug_datasets")
test_dataset_aug_dict.save_to_disk("noun_adj_aug_test_dataset")
```
